# General function to retrieve files in dir

```python
def read_files_from_dir(path, c_dir, format_file=".tsv"):
    from os import listdir
    from os.path import isfile, join

    mypath = path + c_dir

    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

    files_to_open = list(filter(lambda x: format_file in x, onlyfiles))
    
    return files_to_open
```

# General function to read files and concatenate interaction

```python
def create_interactions(path, c_dir, columns, files, name):
    import pandas as pd

    frames = []

    for i in files:
        s = path+c_dir+"\\"+i
        print(s)
        frames.append(pd.read_csv(s, sep="\t", header=0)[columns])

    result = pd.concat(frames)

    result.reset_index(inplace=True)

    result.drop("index", axis=1, inplace=True)

    result.to_csv(path+"\\"+name, sep="\t", header=1)
```
